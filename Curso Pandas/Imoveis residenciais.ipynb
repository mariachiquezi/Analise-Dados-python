# Relatorio de analise 

## imoveis residenciais

```python
#chamando nossos dados 
dados = pd.read_csv('dados/aluguel.csv', sep=';')
```

```python
dados
```

```python
# tirar os duplicados e criar uma lista com os imoveis
list(dados['Tipo'].drop_duplicates())
```

```python
#criando uma nova lista e passando para ela somente os imoveis que classificamos
# como residdencial
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila']
```

```python
residencial
```

```python
# verificar dentro do dataframe se tem algum item da nossa lista reseidencial
# se tiver ele marca como true
selecao=dados['Tipo'].isin(residencial)
```

```python
selecao
```

```python
# manter somente os true
dados_residencial = dados[selecao]
```

```python
dados_residencial
```

```python
#verificação
list(dados_residencial['Tipo'].drop_duplicates())
```

```python
# verificar tamanho
dados_residencial.shape[0]
```

```python
dados.shape[0]
```

```python
#reconstruir indice
dados_residencial.index = range(dados_residencial.shape[0])
```

```python
dados_residencial
```

## exportando a base de dados

```python
# metodo para exportar, estamos criando esse nome de arquivo (aluguel_residencial,csv)
# 
dados_residencial.to_csv('dados/aluguel_residencial.csv', sep=';')
```
