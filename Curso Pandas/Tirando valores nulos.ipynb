## tratamento de dados faltantes (continuação)

```python
dados[dados['Condominio'].isnull()]
```

## assumindo que todo apartamento tem que ter um condominio 
## entao quais nao tiverem condominio sera retirado


```python
# verificando quantos condominios null
dados[dados['Condominio'].isnull()].shape[0]
```

```python
#aqui estamos selecionando quais apartamentos tem condominio
# entao no caso esse selecao contem oq nao queremos remover
selecao = (dados['Tipo']== 'Apartamento') & (dados['Condominio'].isnull())
```

```python
A = dados.shape[0]
# como o selecao nao é oq queremos eliminar precisamos inverter pra pegar o ao
# contrario dele (como se fosse um not in do python), para isso usamos o ~
dados = dados[~selecao]
B = dados.shape[0]
A - B
```

```python
# atribuir o valor 0 aos que restaram 
# dados.fillna(0, inplace = True)
# ou 
# passo a variavel e o valor que eu quero que substitua
dados = dados.fillna({'Condominio': 0 , 'IPTU': 0})
```

```python
# ver quantidades de null
dados[dados['Condominio'].isnull()].shape[0]
```
