# Relatorio Criando variaveis

## criando variaveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep=';')
```

```python
dados.head()
```

```python
# Criando a variavel valor bruto (condominio+iptu+valor)
dados['Valor Bruto'] = dados['Valor']+ dados['Condominio']+ dados['IPTU']
```

```python
dados.head()
```

```python
# valor metro quadrado (valor/area)
dados['Valor m2'] = dados['Valor']/dados['Area']
```

```python
dados
```

```python
# limitando a duas casas decimais
dados['Valor m2'] = dados['Valor m2'].round(2)
```

```python
dados.head()
```

```python
# valor metro quadrado por metro quadrado
dados['Valor Bruto m2'] = (dados['Valor Bruto']/dados['Area']).round(2)
```

```python
dados.head()
```

```python
# pegar o tipo e identificar oq é casa ou apartamento e coloca-las numa variavel
# vamos pegar uma lista com as casass
casa = ['Casa', 'Casa de Condominio', 'Casa de Vila']
```

```python
# aply -> a gente pode aplicar uma função pra cada registro do dataframe
# Tipo vai ser casa se x estiver na variavel casa senao vai ser apartamento
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados
```
