```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15,8))
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
dados.head()
```

```python
# vamsos criar uma area pra colocar o grafico 
area = plt.figure()
```

```python
# queremos 4 graficos
# dentro do nosso figura vamos ter 4 graficos, 2 linhas, 2 colunas e o g1 vai ficar 
# na posição 1 etc
g1 = area.add_subplot(2, 2, 1)
g2 = area.add_subplot(2, 2, 2)
g3 = area.add_subplot(2, 2, 3)
g4 = area.add_subplot(2, 2, 4)
```

```python
# vamos criar um grafico scatter
# com as variaveis valor e area
g1.scatter(dados.Valor, dados.Area)
# dando titulo
g1.set_title('Valor X Area')
g2.hist(dados.Valor)
g2.set_title('Histograma')

# vai selecionar aleatoriamente com limite de 100
dados_g3 = dados.Valor.sample(100)
dados_g3.index = range (dados_g3.shape[0])
g3.plot(dados_g3)
g3.set_title('Amostra (Valor)')

# grafico de barro 
# grupo
grupo = dados.groupby('Tipo')['Valor']
label = grupo.mean().index
valores = grupo.mean().values
g4.bar(label, valores)
g4.set_title('Valor Médio por Tipo')
```
